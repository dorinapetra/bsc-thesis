# Load data

```python
def get_model(row):
    model = "mBERT"
    if row["randomize_embedding_weights"]:
        model = model + "-rand"
        if row["use_character_tokenization"] == "full":
            model = model + "-char"
        #elif row["tokenize_n_chars"] > 0:
         #   model = model + "-lc-" + row["tokenize_n_chars"]
        else:
            model = model + "-subw"
    else:
        if row["use_character_tokenization"] == "full":
            model = model + "-char"
        else:
            model
    return model

def get_model2(row):
    if row["perturbation"] == "mBERT":
        return np.nan
    return row["perturbation"]



exps = pd.read_table("../dataframes/cased_experiments.tsv", sep="\t", quoting=3)

exps["task"] = exps["feature"]
exps["perturbation"] = exps.apply(get_model, axis=1)
#exps["perturbation"] = np.nan
exps["probing_location"] = exps["subword_pooling"]

exps["model"] = "mBERT"

exps = exps[["model", "language", "task", "dev_acc", "test_acc", "perturbation", "probing_location"]]

exps['tag'] = exps['task'].apply(lambda t: t.split("_")[0])
exps['pos'] = exps['task'].apply(lambda t: t.split("_")[1])
exps['full_task'] = exps['language'] + ' ' + exps['task']
exps['family'] = exps['language'].apply(lambda l: lang2family[l])
exps["test_acc"] = exps["test_acc"].apply(lambda a: a * 100)

exps = exps.groupby(["model", "language", "task", "perturbation", "probing_location", "tag","pos","full_task","family"]).mean().reset_index()

exps1 = add_probing_location_better_column(exps)
exps1["perturbation"] = exps1.apply(get_model2, axis=1)

exps1
exps1.perturbation.unique()
```

```python
def get_test_acc(x):
    a = x.split(',')
    a[0] = a[0][1:]
    a[-1] = a[-1][:-1]
    l = map(lambda x: float(x), a)
    return max(l)


def get_model_l2r(row):
    return "mBERT-r2l"



exps = pd.read_table("../dataframes/r2l_ok_experiments.tsv", sep="\t", quoting=3)

exps["task"] = exps["feature"]
exps["probing_location"] = exps["subword_pooling"]
exps["perturbation"] = np.nan

exps["model"] = "mBERT"

exps = exps[["model", "language", "task", "dev_acc", "perturbation", "test_acc", "probing_location"]]

exps['tag'] = exps['task'].apply(lambda t: t.split("_")[0])
exps['pos'] = exps['task'].apply(lambda t: t.split("_")[1])
exps['full_task'] = exps['language'] + ' ' + exps['task']
exps['family'] = exps['language'].apply(lambda l: lang2family[l])
exps["test_acc"] = exps["test_acc"].apply(lambda a: a * 100)
exps["perturbation"] = exps.apply(get_model_l2r, axis=1)

exps = exps.groupby(["model", "language", "task", "perturbation", "probing_location", "tag","pos","full_task","family"]).mean().reset_index()
exps2 = add_probing_location_better_column(exps)


def get_model_l2r_rnd(row):
    return "mBERT-rand-r2l"

exps = pd.read_table("../dataframes/r2l_rnd_ok2_experiments.tsv", sep="\t", quoting=3)
exps["task"] = exps["feature"]
exps["probing_location"] = exps["subword_pooling"]
exps["perturbation"] = np.nan
#exps["test_acc"] = exps.dev_acc_list.apply(get_test_acc)
exps["model"] = "mBERT"

exps = exps[["model", "language", "task", "perturbation", "test_acc", "probing_location"]]

exps['tag'] = exps['task'].apply(lambda t: t.split("_")[0])
exps['pos'] = exps['task'].apply(lambda t: t.split("_")[1])
exps['full_task'] = exps['language'] + ' ' + exps['task']
exps['family'] = exps['language'].apply(lambda l: lang2family[l])
exps["test_acc"] = exps["test_acc"].apply(lambda a: a * 100)
exps["dev_acc"] = 0
exps["perturbation"] = exps.apply(get_model_l2r_rnd, axis=1)

exps = exps.groupby(["model", "language", "task", "perturbation", "probing_location", "tag","pos","full_task","family"]).mean().reset_index()
exps3 = add_probing_location_better_column(exps)

exps3
```

```python
df = exps[
    (exps.model=='mBERT') &
    (exps.probing_location_better) &
    ((exps.perturbation.isin(('mBERT-char', 'mBERT-rand-subw', 'mBERT-rand-char'))) | (exps.perturbation.isnull()))
].copy()
df['perturbation'] = df['perturbation'].fillna('mBERT')
df = df.groupby(['pos', 'language', 'task', 'perturbation'])['test_acc'].mean().reset_index()
df = df[df['language'].isin(('Finnish', 'English', 'German', 'Ukrainian' ,'Slovenian',
                             'Estonian', 'Greek', 'Basque', 'Turkish'))]
df['language'] = df['language'].apply(lambda l: lang2code[l])

fig, ax = plt.subplots(1, figsize=(20, 4))
sns.barplot(x=df['language'], y=df['test_acc'], hue=df['perturbation'], ax=ax)
```

```python
df = exps[
    (exps.model=='mBERT') &
    (exps.probing_location_better) &
    ((exps.perturbation.isin(('mBERT-char', 'mBERT-rand-subw', 'mBERT-rand-char'))) | (exps.perturbation.isnull()))
].copy()
df['perturbation'] = df['perturbation'].fillna('mBERT')
df = df.groupby(['pos', 'language', 'task', 'perturbation'])['test_acc'].mean().unstack().reset_index()
df['diff'] = (1-df['mBERT']) / (1-df['mBERT-char'])
df['rand diff'] = (1-df['mBERT-rand-subw']) / (1-df['mBERT-rand-char'])
df['language'] = df['language'].str.replace('_', ' ')
df['language'] = df['language'].str.replace('Nynorsk', 'N')
df = df.groupby('language').mean()
top_idx = [0, 1, 2, 3, 4, -5, -4, -3, -2, -1]
top1 = df.sort_values('diff').iloc[top_idx]
top2 = df.sort_values('rand diff').iloc[top_idx]

top1 = top1.reset_index()[['language', 'diff']]
top2 = top2.reset_index()[['language', 'rand diff']]
top = pd.concat((top2, top1), axis=1)
top = top.round(3)
lines = top.to_latex(index=False).split('\n')
lines.insert(-8, '\\midrule')
with open(f"{figure_dir}/table_tokenization_diff_minmax.tex", "w") as f:
    f.write("\n".join(lines))
```
